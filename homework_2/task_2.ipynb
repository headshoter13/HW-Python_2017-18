```python
class WikiParser:
    def __init__(self):
        pass
    def get_few_articles(self, start):
        article = Wikipedia().article(start)
        strings = []
        for section in article.sections:
            strings.append(remove_punctuation(' '.join(plaintext(section.source).split(" ")).replace("\n"," ").lower()))
        return strings
    def get_articles(self, start):
        article = Wikipedia().article(start)
        strings = []
        for toPa in article.links:
            temp = Wikipedia().article(toPa)
            for section in temp.sections:
                strings.append(remove_punctuation(' '.join(plaintext(section.source).split(" ")).replace("\n"," ").lower()))
        return strings
```

```python
class TextStatistics:
    def __init__(self, strings):
        self.heart = strings
    def get_top_3grams(self):
        res = defaultdict(int)
        for i in self.heart:
            trigrams = ngrams(i, 3)
            for q in trigrams:
                res[q]+=1
        return list(sorted(res.items(),key=lambda x : -x[1]))
    def get_top_words(self):
        res = defaultdict(int)
        for i in self.heart:
            for q in i.split(" "):
                res[q]+=1
        return list(sorted(res.items(),key=lambda x : -x[1]))
```
